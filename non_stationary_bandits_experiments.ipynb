```python
from MAB import *
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Parameters and aliases for all algorithms

```python
mapping = {'MAG': GaussianMAMAB}
mapping_name = {'MAG': 'Gaussian'}

param = {
    'UCB1': {},
    'weighted_bootstrap': {'gamma': 0.8},
    'weighted_SDA': {'gamma': 0.8},
    'LB_SDA': {'gamma': 0.8},
}

algs = [
    'UCB1', 
    'weighted_bootstrap', 
    'weighted_SDA',
    'LB_SDA',
]  # Select some Algorithms (check param file for availability)
```

### Run experiments

```python
# Settings of the experiments of the paper
xp_gaussian = {
    'xp1': [
        [lambda t: np.cos(2 * np.pi * t / 250), 0.2],
        [lambda t: -np.cos(2 * np.pi * t / 250), 0.2]
    ],
    'xp2': [
        [lambda t: 1.0 * (t > 250) * (t < 750) + (1 - (t > 250) * (t < 750)) * np.cos(2 * np.pi * t / 250), 0.2],
        [lambda t: -1.0 * (t > 250) * (t < 750) - (1 - (t > 250) * (t < 750)) * np.cos(2 * np.pi * t / 250), 0.2],
    ],
    'xp3': [
        [lambda t: t / 100, 0.2],
        [lambda t: -5 + t / 50, 0.2],
        [lambda t: 3 + -t / 250, 0.2],
    ],
    'xp4': [
        [lambda t: -3 + t / 100, 0.2],
        [lambda t: 3 * np.cos(2 * np.pi * t / 100), 0.2],
    ],
}
xp_settings = {'MAG': xp_gaussian}

# General Parameters
xp_family = 'MAG'
T = 1000  # Horizon
N = 1  # Number of MC runs
step = 25  # If results are saved trajectories are stored for all rounds such that t%step=0

xp = [('MAG', 'xp1'), ('MAG', 'xp2'), ('MAG', 'xp3'), ('MAG', 'xp4')]  # To run any subset of xp
for x in xp:
    caption = x[1] + '_' + x[0] + str(int(np.random.uniform() * 1e6))  # Name of the results file
    print(caption)  # caption=None allow to avoid saving the results
    res, traj = multiprocess_MC(
        (x[0], xp_settings[x[0]][x[1]], T, N, algs, param, step, 'mean', False), 
        plot=False,
        caption=False,
    )

    with sns.axes_style('darkgrid'):
        fig, axes = plt.subplots(figsize=(14, 7), nrows=1, ncols=2)
        
        ax = axes[0]
        tt = np.arange(0, T, 0.1)
        for j, arm in enumerate(xp_settings[x[0]][x[1]]):
            ax.plot(tt, arm[0](tt), label='arm {}'.format(j))
        ax.legend(loc='lower right')
        ax.set_title('Reward through time')
        
        ax = axes[1]
        res.plot(ax=ax)
        ax.legend(loc='lower right')
        ax.set_title('Average regret for experiment ' + caption.split('_')[0] + ', ' + mapping_name[x[0]] + ' arms')

        plt.tight_layout()
        plt.show()
```
